```python
#EDA 
# Exploración de valores nulos
# Se revisa la presencia de datos faltantes en las columnas relevantes.

def eda(df):
    print("========== RESUMEN GENERAL ==========")
    print(f"Filas x Columnas (shape): {df.shape}")
    print("\nColumnas:")
    print(df.columns.tolist())

    print("\nDtypes:")
    print(df.dtypes)

    print("\nNulos por columna:")
    print(df.isnull().sum())

    print("\n========== DESCRIBE NUMÉRICO ==========")
    print(df.describe().T)

    print("\n========== DESCRIBE CATEGÓRICO (object/category) ==========")
    print(df.describe(include=["O"]).T)

    print("\n========== HEAD ==========")
    print(df.head())
    print("\n========== TAIL ==========")
    print(df.tail())
    print("\n========== SAMPLE ==========")
    print(df.sample())
    print("\n========== VALUE COUNTS (por columna categórica) ==========")
    col_categoricas =  df.select_dtypes(include=["object", "category"]).columns.tolist()

    for c in col_categoricas:
            print(f"\n--- {c} ---")
            print(df[c].value_counts)

    print("\n========== DUPLICADOS ==========")

    print(df[df.duplicated()])
    if df.duplicated().sum() > 0:
        print("Borrando duplicados:")
        print(df.drop_duplicates())

    print("\n========== HISTOGRAMAS NUMÉRICOS ==========")

    df.hist(bins=20, figsize=(25,25))
    plt.show()

    print("\n========== BARRAS CATEGÓRICAS==========")
    for c in col_categoricas:
         plt.figure(figsize=(10, 10))
         sns.countplot(x=c, data=df, palette='Set2')
         plt.xticks(rotation=90)
         plt.xlabel(f"Countplot de {c}")
         plt.show()
```

# Visualización de datos

Ya tengo el dataframe limpio, así que ahora quiero ver gráficamente la información.

### 1️⃣ ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

Voy a contar cuántos vuelos hay en cada mes.  
Esto me sirve para ver si hay meses con más movimiento o temporadas bajas.


```python
import pandas as pd
import numpy as np
import random
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def vuelos_por_mes(df, mes_col='Month', vuelos_col='Flights Booked'):
    """
    Muestra la cantidad total de vuelos reservados por mes usando groupby().
    """
    vuelos_mes = df.groupby(mes_col)[vuelos_col].sum().reset_index()
    plt.figure(figsize=(8, 5))
    sns.barplot(x=mes_col, y=vuelos_col, data=vuelos_mes, palette='Blues')
    plt.title('Total de Vuelos Reservados por Mes')
    plt.xlabel('Mes')
    plt.ylabel('Total de Vuelos Reservados')
    plt.grid(True)
    plt.show()
    

vuelos_por_mes(df)
```

Los picos indican los meses con más vuelos reservados.

Meses bajos pueden reflejar poca demanda o temporada fuera de vacaciones.

### 2️⃣ ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados?

Tiene sentido pensar que a más distancia, más puntos debería haber.  
Voy a comprobarlo con un diagrama de dispersión.

```python
def distancia_vs_puntos(df, dist_col='Distance', puntos_col='Points Accumulated'):
    """
    Grafica la relación entre distancia y puntos acumulados,
    mostrando el promedio de puntos por distancia con groupby().
    """
    df_group = df.groupby(dist_col)[puntos_col].mean().reset_index()
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df_group, x=dist_col, y=puntos_col, color='darkorange')
    plt.title('Promedio de Puntos Acumulados por Distancia de Vuelo')
    plt.xlabel('Distancia (km)')
    plt.ylabel('Promedio de Puntos')
    plt.grid(True)
    plt.show()

distancia_vs_puntos(df)
```

Una tendencia ascendente sugiere que los vuelos más largos otorgan más puntos.

Si la nube es dispersa, el sistema de puntos puede depender de otros factores.

### 3️⃣ ¿Cuál es la distribución de los clientes por provincia o estado?

Voy a contar cuántos clientes hay en cada estado o provincia.  
Así sé de dónde son la mayoría de los clientes.

```python
def clientes_por_provincia(df, provincia_col='Province'):
    """
    Muestra cuántos clientes hay en cada provincia.
    """
    plt.figure(figsize=(10, 6))
    sns.countplot(y=provincia_col, data=df, order=df[provincia_col].value_counts().index, palette='Set2')
    plt.title('Distribución de Clientes por Provincia')
    plt.xlabel('Cantidad de Clientes')
    plt.ylabel('Provincia')
    plt.show()

clientes_por_provincia(df)
```

Las provincias con más clientes podrían ser mercados más importantes.

Las provincias con pocos clientes podrían representar oportunidades de crecimiento.

### 4️⃣ ¿Cómo se compara el salario promedio entre los diferentes niveles educativos?

Voy a calcular el salario medio según el nivel educativo.  
Esto sirve para ver si hay diferencias claras entre niveles.

```python
def salario_por_educacion(df, edu_col='Education', salario_col='Salary'):
    """
    Compara el salario promedio entre niveles educativos usando groupby().
    """
    edu_salary = df.groupby(edu_col)[salario_col].mean().reset_index()
    plt.figure(figsize=(8, 5))
    sns.barplot(x=edu_col, y=salario_col, data=edu_salary.sort_values(salario_col, ascending=False), palette='coolwarm')
    plt.title('Salario Promedio según Nivel Educativo')
    plt.xlabel('Nivel Educativo')
    plt.ylabel('Salario Promedio')
    plt.xticks(rotation=45)
    plt.show()

salario_por_educacion(df)
```

Niveles educativos más altos tienden a mostrar salarios mayores.

Las diferencias marcan brechas de ingreso según la formación.

### 5️⃣ ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

Voy a ver qué tipo de tarjeta es más común entre los clientes.  
Un gráfico de pastel es perfecto para esto.

```python
def tipos_de_tarjeta(df, tarjeta_col='Loyalty Card'):
    """
    Muestra la proporción de clientes por tipo de tarjeta de fidelidad.
    """
    plt.figure(figsize=(6, 6))
    df[tarjeta_col].value_counts().plot.pie(
        autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel')
    )
    plt.title('Proporción de Tipos de Tarjeta de Fidelidad')
    plt.ylabel('')
    plt.show()

tipos_de_tarjeta(df)
```

Si una tarjeta domina, representa la categoría más común de clientes.

Si hay equilibrio, la base de usuarios está bien distribuida entre niveles.

### 6️⃣ ¿Cómo se distribuyen los clientes según su estado civil y género?

Aquí quiero ver si hay alguna diferencia entre hombres y mujeres según su estado civil.

```python
def estado_civil_y_genero(df, civil_col='Marital Status', genero_col='Gender'):
    """
    Muestra la distribución de clientes según estado civil y género.
    """
    plt.figure(figsize=(8, 5))
    sns.countplot(x=civil_col, hue=genero_col, data=df, palette='Set3')
    plt.title('Distribución de Clientes según Estado Civil y Género')
    plt.xlabel('Estado Civil')
    plt.ylabel('Cantidad de Clientes')
    plt.legend(title='Género')
    plt.show()

estado_civil_y_genero(df)
```

Permite observar si hay predominio de algún género en ciertos estados civiles.

Puede aportar información útil para segmentar campañas de marketing.
